```python
import numpy as np
from sklearn.linear_model import LinearRegression
import warnings
from sklearn.metrics import mean_squared_error

warnings.filterwarnings("ignore")
```

```python
# Dataset
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([52, 55, 61, 70, 82])
```

Model-1

```python
model_1 = LinearRegression()
model_1.fit(X, y)
print("Slope:", model_1.coef_)
print("Intercept:", model_1.intercept_)
```

```python
print("Model_1 Predictions:", model_1.predict(X))
```

```python
x_test = np.array([[6]])
y_pred_test1 = model_1.predict(x_test)
print("Prediction at x=6:", y_pred_test1)
```

```python
mse_1=mean_squared_error(y, model_1.predict(X))
print("Mean Squared Error:", mse_1)
```

Model-2

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([52, 55, 61, 70, 82])

# coefficients
coeff = np.polyfit(x, y, 4)

print(coeff)
```

```python
poly = np.poly1d(coeff)
y_pred_train = poly(x)
print("Predictions on training data:", y_pred_train)
```

```python
y_pred_test2 = poly(6)
print("Prediction at x=6:", y_pred_test2)
```

```python
mse_2=mean_squared_error(y, y_pred_train)
print("Mean Squared Error:", mse_2)
```

Model-1: Higher Bias(Simple Linear Regression) It cannot properly capture the real non-linear pattern between study hours and marks.

Model-2: Higher Variance(Polynomial Regression of degree 4) It fits the training data very closely and is more sensitive to small changes in the data, which can lead to overfitting.
